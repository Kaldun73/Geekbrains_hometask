### 1. К алгоритму kNN, реализованному на уроке, реализовать добавление весов для соседей в зависимости от номера соседа.


```python
import numpy as np
from sklearn import model_selection
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
X, y = load_iris(return_X_y=True)

# Для наглядности возьмем только первые два признака (всего в датасете их 4)
X = X[:, :2]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
cmap = ListedColormap(['red', 'green', 'blue'])
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
```

```python
def e_metrics(x1, x2):
    
    distance = 0
    for i in range(len(x1)):
        distance += np.square(x1[i] - x2[i])
    
    return np.sqrt(distance)
```

```python
def knn_n(x_train, y_train, x_test, k):
    
    answers = []
    for x in x_test:
        test_distances = []
            
        for i in range(len(x_train)):
            
            # расчет расстояния от классифицируемого объекта до
            # объекта обучающей выборки
            distance = e_metrics(x, x_train[i])
            
            # Записываем в список значение расстояния и ответа на объекте обучающей выборки
            test_distances.append((distance, y_train[i]))
        
        # создаем словарь со всеми возможными классами
        classes = {class_item: 0 for class_item in set(y_train)}
        
        # Сортируем список и среди первых k элементов подсчитаем частоту появления разных классов
        i = 1
        for d in sorted(test_distances)[0:k]:
            classes[d[1]] += (k + 1 - i) / k
            i += 1
            
        # Записываем в список ответов наиболее часто встречающийся класс
        answers.append(sorted(classes, key=classes.get)[-1])
    return answers
```

```python
def accuracy(pred, y):
    return (sum(pred == y) / len(y))
```

```python
k = 10

y_pred = knn_n(X_train, y_train, X_test, k)

print(f'Точность алгоритма при k = {k}: {accuracy(y_pred, y_test):.3f}')
```

```python
def get_graph(X_train, y_train, k, knn):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA','#00AAFF'])

    h = .02

    # Расчет пределов графика
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Получим предсказания для всех точек
    Z = knn(X_train, y_train, np.c_[xx.ravel(), yy.ravel()], k)

    # Построим график
    Z = np.array(Z).reshape(xx.shape)
    plt.figure(figsize=(7,7))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Добавим на график обучающую выборку
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f"Трехклассовая kNN классификация при k = {k}")
    plt.show()
```

```python
get_graph(X_train, y_train, k, knn_n)
```

### 2. К алгоритму kNN, реализованному на уроке, реализовать добавление весов для соседей в зависимости от расстояния до соседа. Сравнить метрики качества и сложность разделяющей гиперплоскости. Сделать выводы.


```python
def knn_d(x_train, y_train, x_test, k):
    
    answers = []
    for x in x_test:
        test_distances = []
            
        for i in range(len(x_train)):
            
            # расчет расстояния от классифицируемого объекта до
            # объекта обучающей выборки
            distance = e_metrics(x, x_train[i])
            
            # Записываем в список значение расстояния и ответа на объекте обучающей выборки
            test_distances.append((distance, y_train[i]))
        
        # создаем словарь со всеми возможными классами
        classes = {class_item: 0 for class_item in set(y_train)}
        
        # Сортируем список и среди первых k элементов подсчитаем частоту появления разных классов
        
        for d in sorted(test_distances)[0:k]:
            classes[d[1]] += 0.8**d[0]
            i += 1
            
        # Записываем в список ответов наиболее часто встречающийся класс
        answers.append(sorted(classes, key=classes.get)[-1])
    return answers
```

```python
k = 10

y_pred = knn_d(X_train, y_train, X_test, k)

print(f'Точность алгоритма при k = {k}: {accuracy(y_pred, y_test):.3f}')
```

```python
get_graph(X_train, y_train, k, knn_d)
```

Вывод: точность у обоих алгоритмов оказалась одинакова, но первый на графике выглядит более цельным, без больших оторванных от основной частей. 

### 3. Настроить оптимальное значение числа соседей для моделей из п.1 и п.2, сделать выводы. 


```python
k_list = range(1, 21)
accuracies1 = []
accuracies2 = []
```

```python
for k in k_list:
    y_pred_1 = knn_n(X_train, y_train, X_test, k)
    accuracies1.append(accuracy(y_pred_1, y_test))
    
    y_pred_2 = knn_d(X_train, y_train, X_test, k)
    accuracies2.append(accuracy(y_pred_2, y_test))

print(f'Max accuracy index weights: {max(accuracies1):.3f}')
print(f'Max accuracy distance weights:  {max(accuracies2):.3f}')
plt.plot(k_list, accuracies1, linestyle='dashed', label='weights by index')
plt.plot(k_list, accuracies2, linestyle='dashed', label='weights by distance')
plt.title('Accuracy at k-neighbours increase')
plt.legend(loc='lower right')
plt.show()
```

Вывод: оба алгоритма достигают одного и того же предела качества, но при разных значениях k. Взвешивание по расстоянию дает результат на 5, а по номеру - только на 9.
