#### 1. В чем принципиальные отличия item-item рекомендаций от ALS?


item-item ищет несколько ближайших соседей, которые также оценили этот продукт и на основании них строит рекомендацию. Нужно много вычислений для каждого объекта. 

ALS - рассматривает, как пользователей, так и товары, пытается свести ошибку к минимуму, учитывая оба фактора. Одно вычисление перемножением двух матриц для всех пользователей разом. 

#### 2. Приведите 3 примера весов user-item матрицы для задачи рекоммендаций товаров


* вес подобия двух товаров, если пользователь посмотрел оба
* вес понравившихся товаров
* вес не понравившихся товаров

### 3. Какие ограничения есть у ALS? (Тип информации, линейность/нелинейность факторов и т д)


- нужно обрабатывать строковые значения
- строит рекомендации только по факту взаимодействия

#### 4. Мы рассчматривали bm_25_weight. Опишите, как оно рабтает. Как сделать рекомендации только на основе bm_25? (Можно и нужно пользоваться любыми источниками, приложите на них ссылки)

BM25 обычно используется в ранжировании веб-страниц. его суть заключается в оценке текста на странице, основываясь на количестве и месторасположении ключевых слов, без учета ссылок.
В рекомендациях он взвешивает каждую строку item-user матрицы, основываясь на товарах и выдает максимально релевантные товары в рекомендации
